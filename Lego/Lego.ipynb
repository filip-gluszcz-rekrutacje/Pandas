```python
import pandas as pd
```

```python
lego_set = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
```

```python
themes = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
```

```python
lego_set.head()
```

```python
themes.head()
```

```python
merged = lego_set.merge(themes, left_on='parent_theme', right_on='name')
```

```python
merged.head()
```

```python
merged = merged[~merged['set_num'].isnull()]
```

```python
licensed = merged[merged['is_licensed']]
```

```python
licensed.head()
```

```python
licensed_counts = licensed.groupby('parent_theme')[['name_y']].count()
```

```python
licensed['parent_theme'].value_counts(normalize=True) * 100
```

```python
licensed_counts
```

```python
licensed_counts.loc['Star Wars'] / sum(licensed_counts['name_y']) * 100
```

```python
merged.head()
```

```python
licensed_by_year = merged[merged['is_licensed'] == True]
```

```python
licensed_by_year = licensed_by_year.groupby(['year', 'parent_theme'])[['name_y']].count()
```

```python
licensed_by_year
```

```python
licensed_by_year.reset_index(level='parent_theme')
```

```python
sorted_licensed_by_year = licensed_by_year.sort_values(by='name_y', ascending=False)
```

```python
sorted_licensed_by_year = sorted_licensed_by_year.reset_index(level='year')
```

```python
sorted_licensed_by_year.head(30)
```

```python
sorted_licensed_by_year = sorted_licensed_by_year.drop_duplicates(subset='year', keep='first')
```
